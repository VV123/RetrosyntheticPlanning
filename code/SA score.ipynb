```python
from rdkit import Chem
from rdkit.Chem import RDConfig
from rdkit.Chem import rdMolDescriptors
import pickle
import sys
import time
import math
from collections import defaultdict
import os
import json
import gzip
import six
import pandas as pd
import csv


sys.path.append(os.path.join(RDConfig.RDContribDir, '/content/gdrive/MyDrive/SA_Score/'))
# now you can import sascore!

import sascorer

rows = ['Cc1ccc2c(-c3cc[nH]c(=Nc4ccc(F)c(F)c4)n3)c(-c3ccccc3)nn2n1', 'Cc1ccc2c(-c3ccnc(Nc4ccc(F)c(F)c4)n3)c(-c3ccccc3)nn2n1']

input_variable = []
for row in rows:
    mol = Chem.MolFromSmiles(row)
    sco = sascorer.calculateScore(mol)
    input_variable.append(['%.4f , %s' % (sco, row)])
    print(input_variable)
```
